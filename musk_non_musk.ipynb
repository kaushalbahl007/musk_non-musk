```python
data.drop(["molecule_name","conformation_name"],axis=1,inplace=True)
```

```python
from sklearn.utils import shuffle
```

```python
data=shuffle(data)
```

```python
data.head()
```

```python
from sklearn.model_selection import train_test_split 
```

```python
Y=data['class']
```

```python
Y.shape
```

```python
X=data.drop("class",axis=1)
```

```python
X.shape
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=20,random_state=1)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
```

```python
forest=RandomForestRegressor(random_state=1)
#svm_model=svm.SVC()
kn_model=KNeighborsClassifier(n_neighbors=2)
```

```python
forest.fit(X_train,Y_train)
#svm_model.fit(X_train,Y_train)
kn_model.fit(X_train,Y_train)
```

```python
forest_pred=forest.predict(X_test).round()
kn_pred=kn_model.predict(X_test).round()
```

```python
print(accuracy_score(forest_pred,Y_test))
print(accuracy_score(kn_pred,Y_test))
```

```python
print(confusion_matrix(kn_pred,Y_test))
print(confusion_matrix(forest_pred,Y_test))
```

```python
print(classification_report(Y_test,forest_pred))
print(classification_report(Y_test,kn_pred))
```

```python
import tensorflow as tf
import tensorflow as keras
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D,Activation,MaxPooling2D,Flatten,Dense,MaxPool2D,Dropout
from keras.utils import to_categorical
```

```python
print(X.ndim)
print(X_test.shape[1])
```

```python
#Y_train=to_categorical(Y_train)
#Y_test=to_categorical(Y_test)
```

```python
model=tf.keras.Sequential([
    layers.Dense(64,activation='relu',input_shape=[X.shape[1]]),
    layers.Dense(64,activation='relu'),
    layers.Dense(1)])
optimizer = tf.keras.optimizers.RMSprop(0.001)
model.summary()
```

```python
model.compile(loss='mse',optimizer=optimizer, metrics=['accuracy'])
model.fit(X_train,Y_train, batch_size =20 , epochs =10)
```

```python
test_pred = model.predict(X_test).flatten()
```

```python
print(test_pred.shape)
print(test_pred.ndim)
```

```python
model.evaluate(X_test,  Y_test, verbose=2)
```
